```python
import pandas as pd
topics = pd.read_csv('news topics guardian.csv')
topics_df = pd.DataFrame.from_dict(topics)
topics_df
```

```python
import plotly.express as px
import numpy as np
topics_df["news"] = "news"
fig = px.treemap(topics_df, path=['Topic 0','Topic 1', 'Topic 2', 'Topic 3'], values='Value')
fig.show()
```
